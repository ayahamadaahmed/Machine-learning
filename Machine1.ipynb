```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]
y2 = [1, 2, 3, 4, 5]


plt.subplot(1, 2, 1)  
plt.plot(x, y1, label='y = x^2', color='red')


plt.subplot(1, 2, 2) 
plt.plot(x, y2, label='y = x', color='blue')


plt.tight_layout()
plt.suptitle("Sample Plot")
plt.show()
```

```python
import seaborn as sns
data = sns.load_dataset("tips")
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
sns.lineplot(x = data['total_bill'], y = data['tip'])
plt.title('Total Bill vs Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```

```python
sns.scatterplot(x='total_bill', y='tip', data=data, hue='day')
plt.title('Total Bill vs Tip')
plt.show()
```

```python
sns.lmplot(x='total_bill', y='tip', data=data, hue='day')
plt.title('Total Bill vs Tip')
plt.show()
```

```python
sns.barplot(x='day', y='total_bill', data=data, palette='viridis', ci= None, estimator=np.std)
plt.title('Total Bill by Day')
plt.xlabel('Day of the Week')
plt.ylabel('Total Bill')
plt.show()
```

```python
sns.boxplot(x='day', y='total_bill', data=data, palette='coolwarm')
plt.title('Total Bill Distribution by Day')
plt.xlabel('Day of the Week')
plt.ylabel('Total Bill')
plt.show()
```

```python
numerical_cols = data.select_dtypes(include=['float64', 'int64'])
corr = numerical_cols.corr()
corr
```

```python
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
flights = sns.load_dataset("flights")
flights.head()
```

```python
sns.lineplot(x='year', y='passengers', data=flights)
plt.title('Number of Passengers Over Years')
plt.xlabel('Year')
plt.ylabel('Number of Passengers')
plt.show()
```

```python
sns.histplot(data=flights, x='passengers', bins=20, kde=True)
plt.title('Distribution of Passengers')
plt.xlabel('Number of Passengers')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.boxplot(x='month', y='passengers', data=flights, palette='coolwarm')
plt.title('Passengers by Month')
plt.xlabel('Month')
plt.ylabel('Number of Passengers')
plt.show()
```

```python
# scatter plot relationship between year 1955 and passengers using pandas
import pandas as pd
flights_1955 = flights[flights['year'] == 1955]
plt.scatter(flights_1955['month'], flights_1955['passengers'], color='purple')
plt.title('Passengers in 1955 by Month')
plt.xlabel('Month')
plt.ylabel('Number of Passengers')
plt.show()

```

```python
# create bar chart represent avg monthly passengers per month
avg_passengers = flights.groupby('month')['passengers'].mean().reset_index()
sns.barplot(x='month', y='passengers', data=avg_passengers, palette='viridis')
```
